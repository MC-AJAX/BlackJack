```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
#suits_asci{'Hearts':'\u2665', } #fill this in with each suit asci 
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')

values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

symbols = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':'J', 'Queen':'Q', 'King':'K', 'Ace':'A'}

print_values = {'Two':'00001010000', 'Three':'00001110000', 'Four':'10010001001', 'Five':'10010101001', 'Six':'11010001101', 'Seven':'11010101101', 'Eight':'11011011101', 
            'Nine':'11110101111', 'Ten':'11111011111', 'Jack':'00000J00000', 'Queen':'00000Q00000', 'King':'00000K00000', 'Ace':'00000100000'}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
    
    def print_card(self):
        
        if self.suit == 'Hearts': #pull the asci from the dictionary that matches it to self.suit
            suit_icon = '\u2665'  
        if self.suit == "Diamonds":
            suit_icon = ''
        if self.suit == "Spades":
            suit_icon = ''
        if self.suit == "Clubs":
            suit_icon = ''
        
        
        p_str = print_values[self.rank]#Pulls the associated cards binary interpretation
        symbol = symbols[self.rank]#Pulls the associated cards symbol
        
        p_lst = []
        #Loops through the associated binary interpretation of a card and creates a list of symbols in p_lst
        for i,item in enumerate(p_str):
            
            #adds an empty space to p_lst
            if item == '0':
                p_lst.append(' ')
            
            #Adds the suit symbol to p_lst
            if item == '1':
                p_lst.append(suit_icon)
                
            #For printing Jacks    
            if item == 'J':
                p_lst.append('J')
            
            #For printing Queens
            if item == 'Q':
                p_lst.append('Q')
            
            #For printing Kings
            if item == 'K':
                p_lst.append('K')
        
        #This print function formats the card and pulls either empty spaces or suits from p_lst
        print(f'----------------\n|{symbol}             |\n|   {p_lst[0]}      {p_lst[7]}   |\n|      {p_lst[4]}       |\n|   {p_lst[1]}      {p_lst[8]}   |\n|      {p_lst[5]}       |\n|   {p_lst[2]}      {p_lst[9]}   |\n|      {p_lst[6]}       |\n|   {p_lst[3]}      {p_lst[10]}   |\n|             {symbol}|\n----------------')
```

```python
for item in ranks:
    two_hearts = Card('Hearts',item)
    two_hearts.print_card()
```

```python
class Deck:
        
    def __init__(self):
        # Note this only happens once upon creation of a new Deck
        self.all_cards = [] 
        for suit in suits:
            for rank in ranks:
                # This assumes the Card class has already been defined!
                self.all_cards.append(Card(suit,rank))
                
    def shuffle(self):
        # Note this doesn't return anything
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop()  
```

```python
class Player:
    
    def __init__(self,name,money):
        self.name = name
        # A new player has no cards
        self.all_cards = []
        self.money = money
        self.score = 0
        self.ace_count = 0
        
    def gain_card(self,card):
        #card.print_card()  #once all symbols are added this can be uncommented
        print("Player one recieved: ")
        print(card)
        
        #if the player recieves an ace it increments ace_count by one
        if card.value == 11:
            self.ace_count += 1
            
        self.score = self.score + values[card.rank]
        #Finally print the new score
        print("Player one's score is: ")
        print(self.score)
        
        #returns True if the player busts, false otherwise
        if self.score > 21:
            #Checks if the player has 1 or more Aces and converts them to a score of 1
            if self.ace_count > 0:
                self.score = self.score - 10
                self.ace_count -= 1
                #Finally print the new score
                print("Player one had an Ace, new score is: ")
                print(self.score)
                return "fine"
            self.score = 0
            return "bust"
        else:
            return "fine"


        
    
    
    def __str__(self):
        return f'Player {self.name} has {self.money} in cash'
```

```python
class Dealer: 
    
    def __init__(self):
        self.score = 0
        self.ace_count = 0
              
    def gain_card(self,card):
        #card.print_card()  #once all symbols are added this can be uncommented
        print("The dealer recieved: ")
        print(card)
        
        #if the player recieves an ace it increments ace_count by one
        if card.value == 11:
            self.ace_count += 1
            
        self.score = self.score + values[card.rank]
        print("The dealers score is: ")
        print(self.score)
        #returns fine if the player busts, false otherwise
        if self.score > 21:
            #Checks if the player has 1 or more Aces and converts them to a score of 1
            if self.ace_count > 0:
                self.score = self.score - 10
                self.ace_count -= 1
                #Finally print the new score
                print("Dealer had an Ace, new score is: ")
                print(self.score)
                return "fine"
            self.score = 0
            return "bust"
        else:
            return "fine"
       
    def hit_stay(self):
        #decides if the dealer should hit or stay
        if self.score <= 16:
            return "hit"     
        else: 
            return "stay"
        
```

```python
#Game Logic

player_one = Player("one", 100)
dealer = Dealer()

game_on = True

print("Player one starts with 100")

while game_on == True:
    
    deck_one = Deck()
    deck_one.shuffle()
    
    if player_one.money == 0:
        print("You are out of money")
        game_on = False
        break
    
    #This block of code checks to see if the player has enough money for their bet and if it's an integer
    while True: 
        try:
            p1_bet = int(input("Player One please place your bet: "))
        except:
            print("Please enter an integer")
            continue
        if p1_bet > player_one.money:
            print("You do not have the money to place this bet")
            continue
        #If the bet passes it is reduced from the money that player owns    
        else:
            player_one.money = player_one.money - p1_bet
            break
       
    
    round_num = 0
    round_on = True
    d_round_on = True
    player_one.score = 0
    player_one.ace_count = 0
    dealer.score = 0
    dealer.ace_count = 0
    
    
    #This while loop is for the players turn
    while round_on == True:
        
        p1_choice = "hit" #automatically set to hit for the first 2 rounds
        
        while True:
            #skips this block of code if 2 cards have not deal dealt
            if round_num <= 1:
                break
            #checks if the player entered a correct value
            p1_choice =(input("Please enter 'hit' or 'stay': "))
            p1_choice = p1_choice.lower() #formates the variable as lower case

            if p1_choice == "hit":
                break   
            if p1_choice == "stay":
                break
            else:
                print("Please enter only 'hit' or 'stay': ")
                print(p1_choice)
                continue
                
        
        if p1_choice == "hit":
            #Deals a card and passes it to the player class, result stored in val
            temp_card = deck_one.deal_one()
            val = player_one.gain_card(temp_card)

            #Breaks out of the round_on loop if the player busts
            if val == "bust":
                print("Player one BUST")
                print(player_one)
                d_round_on = False
                break
                
        #Deals the first 2 cards to the dealer
        if round_num <= 1:
            temp_card = deck_one.deal_one()
            val = dealer.gain_card(temp_card)
        
        round_num += 1
        if p1_choice == "stay":  
            break
        
            
            
                
    #This while loop is for the dealers turn     
    while d_round_on == True:
                
        #decides if dealer should hit
        dealer_choice = dealer.hit_stay()
        if dealer_choice == "hit":
            
            #repeats the process for the dealer
            temp_card = deck_one.deal_one()
            val = dealer.gain_card(temp_card)

            #Breaks out of the round_on loop if the dealer busts
            if val == "bust":
                print("Dealer BUST")
                break
                
        else:
            d_round_on = False
            
    if dealer.score >= player_one.score:
        print("Dealer wins")
        while True:
            round_choice = input("Would you like to play again? y or n: ")
            round_choice = round_choice.lower()
            if round_choice == ("y" or "yes"):
                break
            if round_choice == ("n" or "no"):
                game_on = False
                break
            else:
                continue
                
    else:
        player_one.money = player_one.money + (p1_bet * 2)
        print("Player wins")
        print("Player One now has: ")
        print(player_one.money)
        while True:
            round_choice = input("Would you like to play again? y or n: ")
            round_choice = round_choice.lower()
            if round_choice == ("y" or "yes"):
                break
            if round_choice == ("n" or "no"):
                game_on = False
                break
            else:
                continue
        
print("Thanks for playing!!")
```
